# 01 - Carga de datos

En este notebook se realiza la primera comprobacion de los ficheros de MovieLens. El objetivo es verificar que los datos se cargan correctamente y revisar su estructura inicial antes de aplicar cualquier limpieza o transformacion.

## 1. Importacion de librerias

```python
import sys
from pathlib import Path

import pandas as pd
from IPython.display import display

base_dir = Path('..').resolve()
sys.path.append(str(base_dir))

from src.data_utils import (
    load_links,
    load_movies,
    load_ratings,
    load_tags,
    show_basic_info,
)
```

## 2. Definicion de rutas

```python
# Ruta base del proyecto. Desde aqui construiremos el resto de rutas de forma clara y portable.
base_dir = Path('..').resolve()
data_dir = base_dir / 'data' / 'raw'

# Rutas de los archivos originales de MovieLens.
movies_path = data_dir / 'movies.csv'
ratings_path = data_dir / 'ratings.csv'
tags_path = data_dir / 'tags.csv'
links_path = data_dir / 'links.csv'
```

## 3. Carga de los CSV

```python
# Cargamos cada archivo en un DataFrame independiente usando las utilidades del proyecto.
movies_df = load_movies(movies_path)
ratings_df = load_ratings(ratings_path)
tags_df = load_tags(tags_path)
links_df = load_links(links_path)
```

## 4. Vista inicial de los datos

```python
# Mostramos una vista rapida de cada tabla para entender su estructura inicial.
show_basic_info(movies_df, 'movies_df')
print()
show_basic_info(ratings_df, 'ratings_df')
print()
show_basic_info(tags_df, 'tags_df')
print()
show_basic_info(links_df, 'links_df')
```

## 5. Valores nulos

```python
# Comprobamos los valores nulos de forma inicial, sin aplicar limpieza avanzada.
print('movies_df nulls')
display(movies_df.isnull().sum())

print('ratings_df nulls')
display(ratings_df.isnull().sum())

print('tags_df nulls')
display(tags_df.isnull().sum())

print('links_df nulls')
display(links_df.isnull().sum())
```

## 6. Duplicados

```python
# Contamos los registros duplicados para conocer el estado inicial de cada tabla.
print('movies_df duplicated rows:', movies_df.duplicated().sum())
print('ratings_df duplicated rows:', ratings_df.duplicated().sum())
print('tags_df duplicated rows:', tags_df.duplicated().sum())
print('links_df duplicated rows:', links_df.duplicated().sum())
```

## 7. Resumen final

- `movies.csv`: contiene el identificador de la pelicula, el titulo y los generos asociados.
- `ratings.csv`: recoge las valoraciones que los usuarios han dado a las peliculas, junto con la marca temporal.
- `tags.csv`: almacena etiquetas introducidas por usuarios sobre peliculas concretas, tambien con marca temporal.
- `links.csv`: relaciona cada pelicula con identificadores externos de IMDb y TMDb.
